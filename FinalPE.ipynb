```python
user_input = input("Would you like to select Miles Above Mars or Kilometers Above Mars?")
user_number = 0
yards = 0
feet = 0
inches = 0
meters = 0
centimeters = 0
millimeters = 0

if user_input == "Miles Above Mars":
    user_number = float(input("Enter a number"))
    yards = user_number * 1760
    feet = yards * 3  
    inches = feet * 12  
    print(f"Yards: {yards:.2f}\nFeet: {feet:.2f}\nInches: {inches:.2f}")
elif user_input == "Kilometers Above Mars":
    user_number = float(input("Enter a number"))
    meters = user_number * 1000
    centimeters = meters * 100  
    millimeters = centimeters * 10
    print(f"Meters: {meters:.2f}\nCentimeters: {centimeters:.2f}\nMillimeters: {millimeters:.2f}")
```
